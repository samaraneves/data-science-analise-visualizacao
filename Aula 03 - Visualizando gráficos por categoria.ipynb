```python
tmdb = pd.read_csv('C:/Users/sneve/Documents/data-science-analise-visualizacao/tmdb_5000_movies.csv')
```

```python
tmdb.original_language.value_counts()

# Transformando uma Série em um DataFrame
tmdb.original_language.value_counts().to_frame()

# Transformando uma Série em um DataFrame extraindo o index da série, criando assim o index + duas colunas
contagem_linguas = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_linguas.columns = ['original_language', 'total']
contagem_linguas
```

```python
##  Passando uma mensagem através de visualização
```

```python
# Filtrando dados que sejam diferente a determinado dado
idiomas_sem_ingles = tmdb.query("original_language != 'en'")

# Monta um gráfico categórico, usando aspect 2 poderá separar as barras
sns.catplot(x="original_language", kind="count", data=idiomas_sem_ingles, aspect = 2)
```
